```python
def rename_columns(data, old_names, new_names):
    """
    Renames columns in a DataFrame.
    
    Args:
        data (pd.DataFrame): The DataFrame to rename columns in.
        old_names (list): List of old column names.
        new_names (list): List of new column names.
        
    Returns:
        pd.DataFrame: DataFrame with renamed columns.
    """
    
    if len(old_names) != len(new_names):
        raise ValueError("Number of old names and new names should be the same.")
    
    mapping = {old: new for old, new in zip(old_names, new_names)}
    data.rename(columns=mapping, inplace=True)
    
    return data
```

# DATA EXPLORATION

```python
import matplotlib.pyplot as plt
```

```python
years = ['2020/21', '2021/22', '2022/23']
amount_columns = [f'{year} Amount of Funds Allocated (KSh Million)' for year in years]
data_to_plot = data1[['Academic Year'] + amount_columns]

# Plot the trends in the amount offered over the three academic years
data_to_plot.set_index('Academic Year', y=amount_columns, kind='line' inplace=True)
plt.title('Trends in Amount Offered')
plt.xlabel('Academic Year')
plt.ylabel('Amount of Funds Allocated (KSh Million)')
plt.legend(years)
plt.show()
```
